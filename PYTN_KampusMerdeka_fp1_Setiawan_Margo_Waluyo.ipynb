# **Modeling**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=5)
model_LR = LinearRegression()
model = model_LR.fit(x_train, y_train)
ypred = model.predict(x_test)
```

# **EVALUATION**

```python
#cek nilai R2 untuk linear regression
from sklearn import metrics
metrics.r2_score(y_test, ypred)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cross_val = ShuffleSplit(n_splits=5, test_size=0.2, random_state=5)

cross_val_score(LinearRegression(), X, y, cv=cross_val)
```

```python
# hitung akar dari mean squared error (MSE) untuk linear regression
mse = metrics.mean_squared_error(y_test,ypred)
rootmse = np.sqrt(mse)
print(mse)
print(rootmse)
```

```python
mae = metrics.mean_absolute_error(y_test, ypred)
rootmae = np.sqrt(mae)
print(mae)
print(rootmae)
```

# **Model Inference**

Bagian ini diisi dengan model inference, di mana model yang sudah kita latih akan dicoba pada data selain data yang sudah tersedia. Data yang dimaksud bisa berupa data buatan oleh student, ataupun data yang ada pada internet.

```python
def XNew_predict_price(name_cab,distance,surge_multiplier):
    loc_index = np.where(X.columns==name_cab)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = distance
    x[1] = surge_multiplier
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
